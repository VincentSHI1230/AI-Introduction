题目2：利用附件 2 的数据，统计 E151 - E200 这 50 家企业企业进项发票中的有效发票数量，并绘制折线图和柱状图。


```python
import pandas as pd
import matplotlib.pyplot as plt
excel = pd.read_excel('附件2.xlsx', sheet_name='进项发票信息')
valid = excel[excel['发票状态'] == '有效发票']
x = [f'E{i}' for i in range(151, 201)]
y = []
for i in x:
    screen = valid[valid['企业代号'] == i]
    y.append(len(screen))
plt.subplot(2, 1, 1)
plt.plot(range(1, 51), y)
plt.title('x = 1 - 50 : E151 - E200')
plt.subplot(2, 1, 2)
plt.bar(range(1, 51), y)
plt.show()
```
