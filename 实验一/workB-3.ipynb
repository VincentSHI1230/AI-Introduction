题目3：利用附件 3 的数据，拟合出银行贷款年利率与客户流失率关系的线性表达式。

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd
excel = pd.read_excel('附件3.xlsx', sheet_name='Sheet1')
interestRate = excel['贷款年利率'][1:]
lossRate = excel['客户流失率'][1:]
# 拟合线性表达式
def func(x, a, b):
    return a * x + b
popt, pcov = curve_fit(func, interestRate, lossRate)
print(f'拟合得到的线性表达式为: y = {popt[0]:.2f} * x + {popt[1]:.2f}')
# 绘制图像
plt.plot(interestRate, lossRate, 'o', label='original data')
plt.plot(interestRate, func(interestRate, *popt), 'r', label='fitting data')
plt.xlabel('interestRate')
plt.ylabel('lossRate')
plt.legend()
plt.show()
```
