题目1：从键盘输入三个同学的成绩，然后从大到小排列，并且找出最高分。


```python
gradeList = [eval(input(f'请输入成绩{i}: ')) for i in range(3)]
gradeList.sort(reverse=True)
print('从大到小排列为:', gradeList)
print('最高分为:', gradeList[0])
```

题目2：求出 2000 - 2100 的所有闰年，闰年的条件是能被 4 整除，但不能被 100 整除，或者能被 400 整除。


```python
def isRun(year):
    if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0):
        return True
    return False

for i in range(2000, 2101):
    if isRun(i):
        print(i, end=' ')
```

题目3：求 1000 以内最大的 10 个质数的和。


```python
def isPrime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

sumprime = 0
lenprime = 0
i = 1000
while lenprime < 10:
    if isPrime(i):
        sumprime += i
        lenprime += 1
    i -= 1
print(sumprime)
```

题目4：求出 10000 以内的所有完数。如：6 = 1 + 2 + 3，除了它自身以外的因子之和等于它本身叫完数。


```python
def getRealFactor(n):
    factorList = []
    for i in range(1, n):
        if n % i == 0:
            factorList.append(i)
    return factorList

def isPerfect(n):
    if sum(getRealFactor(n)) == n:
        return True
    return False

for i in range(1, 10001):
    if isPerfect(i):
        print(i, end=' ')
```

题目5：生成一个包含 50 个随机整数的列表，然后从后向前删除其中所有的奇数。

```python
import random as r
rlist = [r.randint(1, 100) for i in range(50)]
for i in range(49, -1, -1):
    if rlist[i] % 2 == 1:
        rlist.pop(i)
print(rlist)
```

题目6：令 {a, b, c, d, e, f} 为x轴刻度标签，{1, 2, 3, 4, 5, 6} 为y轴刻度标签。画出 y1 = {5, 3, 2, 4, 1, 6} 和 y2 = {3, 1, 6, 5, 2, 4} 的折线图和柱状图。 


```python
import matplotlib.pyplot as plt
X = ['a', 'b', 'c', 'd', 'e', 'f']
Y1 = [5, 3, 2, 4, 1, 6]
Y2 = [3, 1, 6, 5, 2, 4]
plt.subplot(2, 2, 1)
plt.plot(X, Y1)
plt.ylim(0, 6)
plt.title('Y1')
plt.subplot(2, 2, 2)
plt.plot(X, Y2)
plt.ylim(0, 6)
plt.title('Y2')
plt.subplot(2, 2, 3)
plt.bar(X, Y1)
plt.ylim(0, 6)
plt.subplot(2, 2, 4)
plt.bar(X, Y2)
plt.ylim(0, 6)
plt.show()
```
